```python
from tkinter import *
import json
from tkinter import messagebox


#Complete frame
root=Tk()
root.geometry('200x150') 

#For displaying ideators label at top
topframe=Frame(root)
topframe.pack(side=TOP)
toplabel=Label(topframe,text="Ideators",bg="#A9A9A9", width="200", height="2")
toplabel.pack(side=TOP)

#Frame for buttons
frame=Frame(root)
frame.pack(side=BOTTOM)

#Frame for login and registration form
def createbase():
    root.destroy()
    base=Tk()
    base.geometry('300x200')
    global E1
    global E2
    head=Label(base,text="Please enter user details")
    head.pack(side=TOP)
    L1=Label(base,text="Username:")
    L1.pack(pady=5)
    E1=Entry(base,bd=3)
    E1.pack(pady=5)
    L2=Label(base,text="Password:")
    L2.pack(pady=5)
    E2=Entry(base,bd=3)
    E2.pack(pady=5)
    return base
    
#login verification by reading data stored in text file
def loginverification():
    name=E1.get()
    passw=E2.get()
    json_file=open("userdetails.txt","r")
    user=json.load(json_file)
    stpassw=user.get(name)
    if(stpassw==passw):
        messagebox.showinfo("Message","Login Successful")
    else:
        messagebox.showinfo("Message","Login Failed")
    json_file.close()

#adding new user data in the file in the form of dictionary   
def registeruser():
    name=E1.get()
    passw=E2.get()
    json_file=open("userdetails.txt","r")
    try:
        user=json.load(json_file)
    except:
        user={}
    user[name]=passw
    json_file.close()
    file=open("userdetails.txt","w")
    json.dump(user,file)
    file.close()
    messagebox.showinfo("Message","Successfully Registered")
    
#login form      
def log():
    box=createbase()
    action=Button(box,text="Login",height=2,width=15,command=loginverification)
    action.pack(pady=5)
    box.mainloop()

#registration form
def reg():
    box=createbase()
    action=Button(box,text="Register",height=2,width=15,command=registeruser)
    action.pack(pady=5)
    box.mainloop()

register=Button(frame,text="Register",height=2,width=15,command=reg)
register.pack(pady=5)


login=Button(frame,text="Login",height=2,width=15,command=log)
login.pack(pady=5)


root.mainloop()
```
